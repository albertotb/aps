```python
cmin =  np.inf
cmax = -np.inf
fig, ax = plt.subplots(nrows=len(a_values), ncols=len(d_values), figsize=(20, 120))
avg_cost = np.zeros((len(a_values), len(d_values)))
for i, d in enumerate(d_values):
    for j, a in enumerate(a_values):
        l_values = prob(d, a, size=1000000)
        cost = ca(a, l_values, 0.002, 0.0026, 0.00417)
        ax[j, i].hist(l_values, density=True, align='left')
        ax[j, i].set_title('{}'.format(l_values.mean()))
        avg_cost[j, i] = ua(cost, ka=9).mean()
        lmin, lmax = cost.min(), cost.max()
        if lmin <= cmin:
            cmin = lmin
        if lmax >= cmax:
            cmax = lmax
```

```python
pd.set_option('precision', 12)
pd.DataFrame(avg_cost, index=a_values, columns=d_values)
```
